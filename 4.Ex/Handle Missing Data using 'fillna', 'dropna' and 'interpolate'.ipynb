**Convert string column into the date type**


```python
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace = True)
df
```

## 4.1 Use fillna() method in dataframe

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method = 'ffill') #ffill means forward fill(similar is 'bfill')
new_df
```

**axis" parameter in fillna() method in dataframe**

```python
new_df = df.fillna(method = 'bfill', axis = 'columns') # horizontal copied
new_df
```

**"limit" parameter in fillna() method in dataframe**

```python
new_df = df.fillna(method='ffill', limit= 1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method = 'time')
new_df
```

## 4.3  dropna() method Drop all the rows which has "na" in dataframe

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how = 'all')
new_df
```

```python
new_df = df.dropna(thresh=2) #return rows has at least 2 valid values
new_df
```

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
