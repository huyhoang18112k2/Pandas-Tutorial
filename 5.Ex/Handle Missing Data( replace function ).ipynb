# 5. Handle Missing Data: replace function

```python
%autosave 60
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.replace(['-99999','-88888'], np.NaN)
new_df
```

```python
new_df1 = df.replace({
    'temperature': '-99999',
    'windspeed': ['-99999','-88888'],
    'event': '0'
}, np.NaN)
new_df1
```

```python
new_df2  =df.replace({
    '-99999': np.NaN,
    '-88888': np.NaN,
    '0': 'Hellooo'
})
new_df2
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},'',regex = True)
new_df.windspeed = new_df.windspeed.apply(lambda x: float(x))
new_df
```

```python
type(new_df.windspeed[0])
```

**Replace list with another list**
