#  Crosstab

**Crosstab is a type of table in a matrix format  that displays the (multivariate) frequency distribution of the variables**

```python
%autosave 60
```

```python
import pandas as pd
import numpy as np
df = pd.read_excel('survey.xls' )
df
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
pd.crosstab(df.Sex, [df.Handedness, df.Nationality] ,margins=True)
```

```python
pd.crosstab([df.Sex, df.Name], [df.Handedness, df.Nationality])
```

```python
pd.crosstab(df.Sex, [df.Handedness] ,margins=True)
```

```python
pd.crosstab(df.Sex, [df.Handedness] ,normalize='index')
```

```python
pd.crosstab(df.Sex, [df.Handedness], values=df.Age, aggfunc=np.mean )
```
